````python
def show_image_details(img_file_name):
  """
  1. You get a NumPy array with the following structure: shape = (height, width, channels)
  X rows     → image height
  Y columns  → image width
  Z channels → B, G, R or None(Garyscale)

  2. Datatype of each pixel
  dtype = uint8 = 2⁸ = 256 values
  0   → black/no intensity
  255 → full intensity
  """
  image = cv2.imread(img_file_name)
  print("Image Funct: ", dir(image))
  print("Image shape: ", image.shape)
  print("Image Size : ", image.size)
  print("==========================")
  return image
````

````python
def show_image(img, title="Dogs"):
    """
    1. Colors will be wrong (blues look orange, etc.)
    plt.imshow(img)

    2. OpenCV uses BGR, Matplotlib uses RGB
    rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(rgb)
    """
    plt.figure(figsize=(12, 12))
    plt.figure(figsize=(8, 8))
    if len(img.shape) == 3:

      plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    else:  # Grayscale image
      # cmap (colormap for grayscale images)
      # Without this, grayscale will appear as a rainbow (viridis colormap).
      plt.imshow(img, cmap='gray')
      # plt.imshow(img, cmap='gray', vmin=0, vmax=255)
    plt.title(title)
    plt.axis("off")
    plt.show()
````

### RGB - Understand Image Matrix Output

When you print a color image, you see a 3D array structure:
```
[[ 56  95  74]  ← One pixel
  [ 56  95  74]  ← Another pixel
  ...
]  ← One row of pixels
```

### What Each Pixel Represents

Each set of **three numbers** `[56 95 74]` represents **one pixel** with three color channels:

### BGR Format (OpenCV's default):
- **First number (56)**: Blue channel intensity
- **Second number (95)**: Green channel intensity  
- **Third number (74)**: Red channel intensity

### Value Range:
Each value ranges from **0-255**:
- **0** = no color (darkest)
- **255** = full color (brightest)

### Complete 3D Array Structure:
```
3D Array Shape: (Height, Width, 3)
├── Dimension 1: Rows (height)
├── Dimension 2: Columns (width)
└── Dimension 3: Color channels [B, G, R]
```

### Example Analysis:
For pixel `[56, 95, 74]`:
- Low blue (56)
- Medium green (95)
- Medium-low red (74)
- Result: **Greenish color**

**Note:** OpenCV uses **BGR** (not RGB)! This is why we need `cv2.cvtColor(img, cv2.COLOR_BGR2RGB)` when displaying with matplotlib.

### GYS - Understanding Grayscale Matrix Output

When you print a grayscale image, you see a 2D array structure:
```
[[81 81 81 ... 79 79 79]  ← One row of pixels
 [81 81 81 ... 79 79 79]  ← Another row
 [81 81 81 ... 79 79 79]
 ...
 [77 77 77 ... 86 86 86]]
```

### What Each Number Represents

Each **single number** represents **one pixel's intensity**:

- `81` = One pixel with intensity 81
- `79` = One pixel with intensity 79
- `86` = One pixel with intensity 86

### Grayscale Format:
- **Only ONE channel** (no separate R, G, B)
- Each value represents **brightness/intensity**

### Value Range:
Each value ranges from **0-255**:
- **0** = Black (darkest)
- **255** = White (brightest)
- **128** = Medium gray
- Other values = Various shades of gray

### Complete 2D Array Structure:
```
2D Array Shape: (Height, Width)
├── Dimension 1: Rows (height)
└── Dimension 2: Columns (width)
```

### Comparison:

| Image Type | Array Shape | Values per Pixel |
|------------|-------------|------------------|
| **Color (BGR)** | (Height, Width, 3) | `[B, G, R]` - 3 numbers |
| **Grayscale** | (Height, Width) | Single number |

### Example Analysis:
- Pixel value `81` = Dark gray (closer to black)
- Pixel value `79` = Slightly darker gray
- Pixel value `86` = Slightly lighter gray

**Note:** Grayscale images are simpler (2D) compared to color images (3D), making them faster to process!

## Image Flipping
```
flipCode	Flip Direction  Description
1	        Horizontal      Mirror left ↔ right
0	        Vertical        Flip top ↔ bottom
-1	        Both	        Rotate 180° (H + V flip)

horizontal = cv2.flip(img, 1) # Left ↔ Right
Original:       Flipped:
A B C           C B A
D E F    →      F E D
G H I           I H G

vertical = cv2.flip(img, 0) # Top ↔ Bottom
Original:       Flipped:
A B C           G H I
D E F    →      D E F
G H I           A B C

both = cv2.flip(img, -1) # Both Horizontal + Vertical (180° rotation)
Original:       Flipped:
A B C           I H G
D E F    →      F E D
G H I           C B A
```

## HSV


### HSV Color Model - Table Format

| Component | Known as | Abbreviation | Range | Description | Controls |
|-----------|----------|--------------|-------|-------------|----------|
| **H - Hue** | **Color** | H | 0° - 360° | Pure color type | What color it is |
| **S - Saturation** | **Color Intensity** | S | 0% - 100% | Color intensity/purity | How vivid the color is |
| **V - Value** | **Color Brightness** | V | 0% - 100% | Brightness level | How bright/dark it is |

---

### What Happens When You Change HSV Values

| Change | Effect | Example |
|--------|--------|---------|
| Increase Hue | Different color | Red → Yellow → Green → Blue |
| Decrease Hue | Different color (reverse) | Blue → Green → Yellow → Red |
| Increase Saturation | More vivid/pure | Pastel → Vibrant |
| Decrease Saturation | More washed out | Vibrant → Gray |
| Increase Value | Brighter | Dark → Light |
| Decrease Value | Darker | Light → Black |

---

### HSV vs Natural Language

| You Say | HSV Adjustment |
|---------|----------------|
| "I want blue" | Set Hue = 240° |
| "Make it more vivid" | Increase Saturation |
| "Make it pastel" | Decrease Saturation |
| "Make it darker" | Decrease Value |
| "Make it brighter" | Increase Value |
| "Remove the color" | Set Saturation = 0% |

---

### Special Cases

| Condition | Result | Why |
|-----------|--------|-----|
| V = 0% | Always black | No brightness = no light |
| S = 0% | Always gray | No saturation = no color |
| S = 0%, V = 100% | White | Brightest gray possible |
| S = 0%, V = 50% | Medium gray | No color, medium brightness |
| S = 100%, V = 100% | Pure color | Maximum intensity and brightness |
